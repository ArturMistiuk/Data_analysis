```python
import pandas as pd
import numpy as np
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"
tmp = pd.read_html(url, match="Коефіцієнт народжуваності в регіонах України", thousands=".")
tmp = tmp[0]
print(tmp.head())
```

```python
print(tmp.shape)
```

```python
tmp.replace("—", np.nan, inplace=True)
print(tmp)
```

```python
print(tmp.dtypes)
```

```python
tmp = tmp.fillna(0)
print(tmp)
```

```python
print(tmp.isnull().sum())
```

```python
print(tmp.drop([27]))
```

```python
tmp_region = tmp.iloc[0:, :1]
tmp = tmp.iloc[0:, 1:]
print(tmp)
```

```python
tmp = tmp.replace(0, np.nan)
tmp = tmp.replace(',', '.', regex=True).astype(float)
tmp = tmp.fillna(tmp.mean()).round(1)
print(tmp)
```

```python
tmp.insert(0, "Регiон", tmp_region)
print(tmp)
```

```python
average_birth_rate_2019_UA = tmp.iloc[-1].iloc[-1]
high_birth_rate_regions = tmp[tmp["2019"] > average_birth_rate_2019_UA]["Регiон"]
print(high_birth_rate_regions)
```

```python
max_birth_rate_2014 = tmp[tmp["2014"] == tmp["2014"].max()][['Регiон', '2014']]
print(max_birth_rate_2014)
```

```python
import matplotlib.pyplot as plt


plt.bar(
    tmp["Регiон"],
    tmp["2019"],
    color=['#FF6B6B', '#6BFFB8', '#6B96FF', '#FFE66B', '#FF6BD0'],
       )

plt.xlabel("Region", fontsize="small", color='#1F618D')
plt.ylabel("Birth Rate", fontsize="small", color='#1F618D')
plt.title("Birth rate in Ukraine in 2019", fontsize=15)
plt.xticks(rotation=90) 
plt.show()
```

```python
zp_region = tmp.loc[tmp['Регiон'] == 'Запорізька']
years = tmp.columns[1:]
```

```python
plt.plot(years,
        zp_region.iloc[0, 1:],
        label='Zaporizhzhya birth rate',
        color="#FF5733",
        linewidth=2,
         marker="o"
        )
plt.xlabel("Рiк", fontsize="small", color="midnightblue")
plt.ylabel("Народжуванiсть", fontsize="small", color="midnightblue")
plt.title("Народжуванiсть У Запорiжжi", fontsize=15)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(tmp["Регiон"], tmp["2014"], color='blue', label='2014')
plt.xlabel("Регіон")
plt.ylabel("Народжуваність")
plt.title("Зміна народжуваності за регіонами в 2014 році")
plt.legend()
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.fill_between(tmp.columns[1:], tmp.loc[tmp["Регiон"] == "Україна", "1950":"2019"].values[0], color='green', alpha=0.5)
plt.xlabel("Рік")
plt.ylabel("Народжуваність")
plt.title("Зміна народжуваності в Україні з 1950 по 2019 роки")
plt.tight_layout()

plt.show()
```
